```python
from qiskit import *
import numpy as np
from math import sqrt, pi
from qiskit.visualization import plot_histogram, plot_bloch_multivector
%matplotlib inline
```

```python
#Recall the bloch sphere state vector equation
'''|q> = cos(theta/2)|0> + e^(i*phi)sin(phi/2)|1>'''
```

## The Pauli Gates

### The X Gate

```python
#x is the pauli x matrix
#Flips amplitudes of 0 and 1 basis
qc = QuantumCircuit(1)
qc.x(0)
qc.draw(output='mpl')
```

```python
#qc.save_statevector()
sim = Aer.get_backend('statevector_simulator')
final_state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(final_state)
```

```python
#same for z and y
qc = QuantumCircuit(1)
qc.x(0)
qc.y(0)
qc.z(0)
qc.draw(output='mpl')
```

```python
final_state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(final_state)
```

### Measuring in the X basis, rather than the Z basis

```python
#Function that measures in the x-basis
def x_measurement(qc, qubit, cbit):
    """Measure 'qubit' in the X-basis, and store the result in 'cbit'"""
    qc.h(qubit)
    qc.measure(qubit, cbit)
    return qc

initial_state = [1/sqrt(2), -1/sqrt(2)]
# Initialize our qubit and measure it
qc = QuantumCircuit(1,1)
qc.initialize(initial_state, 0)
x_measurement(qc, 0, 0)  # measure qubit 0 to classical bit 0
qc.draw(output='mpl')
```

```python
counts = sim.run(qc).result().get_counts()
plot_histogram(counts)
```

### Phase gates

```python
initial_state = [1/sqrt(2), 1/sqrt(2)]
#initial_state = [1,0]
# Initialize our qubit and measure it
qc = QuantumCircuit(1)
qc.initialize(initial_state, 0)

qc.p(pi/4, 0)
qc.h(0)

qc.save_statevector()
qc.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
counts = sim.run(qc).result().get_counts()
plot_histogram(counts)
```

```python
initial_state = [1/sqrt(2), 1/sqrt(2)]
#initial_state = [1,0]
# Initialize our qubit and measure it
qc = QuantumCircuit(1)
qc.initialize(initial_state, 0)

qc.s(0)
qc.sdg(0) #inverse s gate (not it's own inverse!)

qc.save_statevector()
qc.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()
counts = result.get_counts()
plot_bloch_multivector(result.get_statevector())
```

```python
initial_state = [sqrt(2/3), sqrt(1/3)]
#initial_state = [1,0]
# Initialize our qubit and measure it
qc = QuantumCircuit(1)
qc.initialize(initial_state, 0)

qc.t(0)
#qc.tdg(0) #inverse t gate (not it's own inverse!)

qc.save_statevector()
qc.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()
counts = result.get_counts()
plot_bloch_multivector(result.get_statevector())
```

### The U Gate (the most general gate)

```python
qc = QuantumCircuit(1)


qc.h(0)
qc.u(pi,pi/2,pi,0)

qc.save_statevector()
qc.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()
counts = result.get_counts()
plot_bloch_multivector(result.get_statevector())
```
